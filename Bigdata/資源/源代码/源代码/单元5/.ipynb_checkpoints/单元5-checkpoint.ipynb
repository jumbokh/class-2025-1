## 代码5-4

```python
import pandas as pd
import numpy as np
#随机产生三行两列的数据集
df = pd.DataFrame(np.random.randn(3,2), index = list('abc'),columns=['1','2'])
#将部分指定位置定义为缺失
df.iloc[1:-1,1] = np.nan
print(df)
```

```python
#利用dropna 函数删除含有缺失值的行
df.dropna()
```
