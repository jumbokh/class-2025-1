## 代码7-1

```python
import pandas as pd 
df = pd.DataFrame({"key":['张三','李四','王五','王五','李四','张三'],
                  "data":[2,4,6,8,10,13]})
print(df)
```

## 代码7-2

```python
#按照“key”列数据进行分组，使用列索引字段
group1 = df.groupby(by='key')
#遍历查看
for item in group1:
    print(item)
```

## 代码7-3

```python
#按Series 对象分组
df = pd.DataFrame({"key1":['张三','张三','李四','李四','张三'],
"key2":['one','two','one','two','one'],
"data1":[2,3,4,6,8],
"data2":[3,5,6,3,7]})
print(df)
```

## 代码7-4

```python
se = pd.Series(['a','b','a','c','b'])
group2 = df.groupby(by=se)
for item in group2:
    print(item)
```

## 代码7-5

```python
#Series 对象与pandas 对象的索引长度不相同
se = pd.Series(['a','b','a'])
group2 = df.groupby(by=se)
for item in group2:
    print(item)
```

## 代码7-6

```python
#按字典分组
df = pd.DataFrame({"a":[1,3,5,7,9],
"b":[2,3,4,6,8],
"c":[5,4,3,2,1],
"d":[3,5,6,3,7]})
#定义映射关系
mapping = {'a':'一组','b':'二组','c':'二组','d':'一组'}
group3 = df.groupby(by=mapping, axis=1)
for item in group3:
    print(item)
```

## 代码7-7

```python
#按函数分组
df = pd.DataFrame({"a":[1,3,5],
"b":[2,3,4],
"c":[5,4,3]},
index=['张三','李四六','王五七'])
print(df)
```

## 代码7-8

```python
#按照长度函数分组
#传递长度len 函数
group4 = df.groupby(len)
for item in group4:
    print(item)
```

## 代码7-18

```python
def subtract_and_divide(x, sub, divide=1):
    return (x - sub) / divide
data.apply(subtract_and_divide, args=(5,), divide=3)
```

## 代码7-19

```python
def subtract_and_divide(x, sub, divide=1):
    return (x - sub) / divide
data.loc[:,['a', 'b']].apply(subtract_and_divide, args=(5,), divide=3)
```
